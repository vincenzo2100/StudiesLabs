```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

Wczytujemy plik diabetes.txt,info i określamy listę atrybutów

```python
attribute_list = ["a1n", "a2n", "a3n", "a4n", "a5n", "a6n", "a7n", "a8n", "a9d"]
data = pd.read_csv("diabetes.txt", sep=" ", header=None)
info = pd.read_csv("_info-data-discrete.txt", sep=" ", header=None)
data
```

a)
Szukamy brakujących wartości atrybutów

```python
num_missing = (data[[0, 1, 2, 3, 4, 5, 6, 7]] == 0).sum()
print(num_missing)
```

Wypełniamy brakujące wartości NaN(lepiej wypełnić niżeli znakiem "?")

```python

data[[0, 1, 2, 3, 4, 5, 6, 7]] = data[[0, 1, 2, 3, 4, 5, 6, 7]].replace(0, np.nan)
data
```

Naprawiamy metodą średniej

```python
data.fillna(data.mean(), inplace=True)
data
```
